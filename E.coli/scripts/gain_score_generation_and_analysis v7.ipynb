```python
def matrix_completion_fcgr(fcgr_sequence_ecoli):

    import numpy as np
    import kaos

    kmer_length = 10

    chaos_game_kmer_array = kaos.kaos_fcgr().chaos_game_representation_key(kmer_length=kmer_length)



    masked_array = np.where(fcgr_sequence_ecoli <= 3, np.nan, fcgr_sequence_ecoli)
    masked_array = np.nan_to_num(masked_array, nan=0)
    mask = (masked_array > 0).astype(int)
    alpha= 0.00002
    beta=0.002
    K =150
    row, col = 1024, 1024
    user_matrix= np.random.rand(row, K)
    item_matrix = np.random.rand(K, col)
    steps =100
    # Perform gradient descent
    for i in range(steps):
        
        # Compute the error matrix E
        E = (masked_array - np.dot(user_matrix, item_matrix)) * mask
        # print(E)

        # print("NMAE Training error: ", np.mean(np.abs(E))/4)
        
        # Compute the gradient
        grad_P = -2 * np.dot(E, item_matrix.T) + 2 * beta * user_matrix
        grad_Q = -2 * np.dot(user_matrix.T, E) + 2 * beta * item_matrix
        
        # Update P and Q using gradient descent
        user_matrix -= alpha * grad_P
        item_matrix -= alpha * grad_Q

    predicted_ratings = np.dot(user_matrix, item_matrix)

    df1 = pd.DataFrame(columns = ["kmer", "freq"])
    df1["kmer"] = chaos_game_kmer_array.ravel()
    df1["freq"] = predicted_ratings.ravel()

    result_dict = dict(zip(df1.iloc[:, 0], df1.iloc[:, 1]))

    

    return predicted_ratings, result_dict


```

#### Generate gain score

```python
df_sliding = pd.DataFrame(columns = ['Position', 'Gene', 'Allele', 'Annotation', 'Test.statistic', 'P-value',
       'Ref_allele', 'Alt_allele', 'ref_flanking_regions',
       'alt_flanking_regions', 'label'])
df_complete = pd.DataFrame(columns = ['Position', 'Gene', 'Allele', 'Annotation', 'Test.statistic', 'P-value',
       'Ref_allele', 'Alt_allele', 'ref_flanking_regions',
       'alt_flanking_regions', 'label'])


for i in range(df.shape[0]):
    
    ref_seq = df.loc[i, "ref_flanking_regions"]
    alt_seq = df.loc[i, "alt_flanking_regions"]
    
    center_pos = int(len(ref_seq)/2)
    
    gain = 0
    
    for j in range(kmer_length):
        
       #  print(j)

        df_sliding.loc[j, "altered sliding window"] = alt_seq[center_pos - kmer_length+j+1:center_pos+1+j]
        df_sliding.loc[j, "reference sliding window"] = ref_seq[center_pos - kmer_length+j+1:center_pos+j+1]

        df_sliding.loc[j, "gain"] = np.log(fcgr_sequence_Ecoli[df_sliding.loc[j, "altered sliding window"]]/ fcgr_sequence_Ecoli[df_sliding.loc[j, "reference sliding window"]])
    
        gain = gain + df_sliding.loc[j, "gain"]
    
    df_sliding['Position'] = df.loc[i, 'Position']
    df_sliding['Gene'] = df.loc[i, 'Gene']
    df_sliding['Allele'] = df.loc[i, 'Allele']
    df_sliding['Annotation'] = df.loc[i, 'Annotation']
    df_sliding['Test.statistic'] = df.loc[i, 'Test.statistic']
    df_sliding['P-value'] = df.loc[i, 'P-value']
    df_sliding['Ref_allele'] = df.loc[i, 'Ref_allele']
    df_sliding['Alt_allele'] = df.loc[i, 'Alt_allele']
    df_sliding['ref_flanking_regions'] = df.loc[i, 'ref_flanking_regions']
    df_sliding['alt_flanking_regions'] = df.loc[i, 'alt_flanking_regions']
    df_sliding['label'] = df.loc[i, 'label']
    df_sliding["accumulated_gain"] = gain

    df_complete = pd.concat([df_complete, df_sliding])
```

##### Find gene wise p value for median esm and median AG score

###### For each gene, we compute the median AG score and median ESM scores with respect to every available generation. Subsequently, the median score for each gene is partitioned into two groups based on a 30,000-generation threshold. The medians score for generations less than or equal to 30,000 are grouped together, while those for generations greater than 30,000 are placed in another group for both AG and ESM. Reporting for a gene is omitted if either of the two scores has fewer than or equal to 15 samples in any of the corresponding groups. Subsequently, for each score, we calculate a one-sided Mann-Whitney test score between the two aforementioned groups. This methodology is iteratively applied to each gene.

##### calculate p value

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import mannwhitneyu

# Separate the data into mutator and non-mutator groups
mutator_data = df_gene_type_gain[df_gene_type_gain['mutator'] == 'mutator']
non_mutator_data = df_gene_type_gain[df_gene_type_gain['mutator'] == 'non mutator']

# Create a function to perform Mann-Whitney U test and return the test statistic and p-value
def perform_mannwhitneyu(data1, data2):
    stat, p_value = mannwhitneyu(data1, data2, alternative='greater')
    return stat, p_value

# Perform Mann-Whitney U test for mutator group
mutator_AG_essential = mutator_data[mutator_data['type'] == 'essential ']['AG']
mutator_AG_non_essential = mutator_data[mutator_data['type'] == 'non-essential ']['AG']

test_stat_mutator, p_value_mutator = perform_mannwhitneyu(mutator_AG_essential, mutator_AG_non_essential)
print(f"Mutator Group: Mann-Whitney U Test Statistic = {test_stat_mutator}, p-value = {p_value_mutator}")

# Perform Mann-Whitney U test for non-mutator group
non_mutator_AG_essential = non_mutator_data[non_mutator_data['type'] == 'essential ']['AG']
non_mutator_AG_non_essential = non_mutator_data[non_mutator_data['type'] == 'non-essential ']['AG']

test_stat_non_mutator, p_value_non_mutator = perform_mannwhitneyu(non_mutator_AG_essential, non_mutator_AG_non_essential)
print(f"Non-Mutator Group: Mann-Whitney U Test Statistic = {test_stat_non_mutator}, p-value = {p_value_non_mutator}")
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import mannwhitneyu

# Assuming 'all_mutation' and 'mutator_list' are defined

# Add a column for citrus_utilization_gen based on the condition
all_mutation["citrus_utilization_gen"] = np.where(all_mutation['generation_number'] < 33500, "<33500", '>=33500')

# Create a function to perform Mann-Whitney U test and return the test statistic and p-value
def perform_mannwhitneyu(data1, data2):
    stat, p_value = mannwhitneyu(data1, data2, alternative='greater')
    return stat, p_value

# Perform Mann-Whitney U test for each label separately and print the test statistics and p-values
for label in mutator_list:
    data_low = all_mutation[(all_mutation["label"] == label) & (all_mutation['citrus_utilization_gen'] == '<33500')]['AG']
    data_high = all_mutation[(all_mutation["label"] == label) & (all_mutation['citrus_utilization_gen'] == '>=33500')]['AG']
    
    test_stat, p_value = perform_mannwhitneyu(data_high, data_low)  # Switched data1 and data2
    
    # Print the Mann-Whitney U test statistic and p-value for each label
    print(f"{label}: Mann-Whitney U Test Statistic = {test_stat}, p-value = {p_value}")
```

```python

def plot_mutator_non_mutator_graph(ax, data_pivot_median, data_pivot_std, mutator_list, non_mutator_list, typeplot,offset=0):

    data_pivot_median = data_pivot_median.copy()
    data_pivot_std = data_pivot_std.copy()
    c_index_mutator = 0
    c_index_non_mutator = 0

    generation_number_median = data_pivot_median.index
    generation_number_std= data_pivot_std.index

    if np.array_equal(generation_number_median, generation_number_std)!= True:

        print("Generation mismatch in std and median dataframe")
        exit

    for i in data_pivot_median.columns:

        generation_wise_median = data_pivot_median[i]
        generation_wise_std = data_pivot_std[i]

        if i in mutator_list:

            ax[c_index_mutator, 0].plot(generation_number_median,  generation_wise_median, lw = 0.5, c = "red", label = "median")
            ax[c_index_mutator, 0].fill_between(generation_number_median, (generation_wise_median-generation_wise_std), (generation_wise_median+generation_wise_std), color='lime', alpha=0.2)
            ax[c_index_mutator, 0].set_ylabel(typeplot + " of " + i , fontsize= 2)
            # ax[c_index_mutator, 0].set_xlim(0, 2020)
            ax[c_index_mutator, 0] = utility.custom_figure_axis(ax[c_index_mutator, 0], fontsize=2, show_ticks = True)


            spearman_corr, spearman_p = utility.spearmanr(generation_number_median, generation_wise_median)
            # print(offset)

            ax[c_index_mutator, 0].text((max(ax[c_index_mutator, 0].get_xlim()) -100* offset), (max(ax[c_index_mutator, 0].get_ylim()) - offset), "Corr:" + str(spearman_corr) + "| p- value: " + str(spearman_p), fontsize=2, color='red', ha='right', va='top')

            spearman_corr = None
            spearman_p = None
            c_index_mutator = c_index_mutator + 1

        if i in non_mutator_list:

            ax[c_index_non_mutator, 1].plot(generation_number_median,  generation_wise_median, lw = 0.5, c = "blue", label = "median")
            ax[c_index_non_mutator, 1].fill_between(generation_number_median, (generation_wise_median-generation_wise_std), (generation_wise_median+generation_wise_std), color='lime', alpha=0.2)
            ax[c_index_non_mutator, 1].set_ylabel(typeplot + " of " + i , fontsize= 2)
            # ax[c_index_non_mutator, 0].set_xlim(0, 2020)
            ax[c_index_non_mutator, 1] = utility.custom_figure_axis(ax[c_index_non_mutator, 1], fontsize=2, show_ticks = True)


            spearman_corr, spearman_p = utility.spearmanr(generation_number_median, generation_wise_median)
            ax[c_index_non_mutator, 1].text((max(ax[c_index_non_mutator, 1].get_xlim()) -100* offset), (max(ax[c_index_non_mutator, 1].get_ylim()) - offset), "Corr:" + str(spearman_corr) + "| p- value: " + str(spearman_p), fontsize=2, color='red', ha='right', va='top')

            spearman_corr = None
            spearman_p = None

            c_index_non_mutator = c_index_non_mutator + 1

        del generation_wise_median, generation_wise_std


    return ax

        
```
