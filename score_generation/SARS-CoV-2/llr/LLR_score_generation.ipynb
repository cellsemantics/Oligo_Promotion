```python
import pandas as pd
# read the df 
data = pd.read_excel("../esm1b_analysis/orf1ab_10/df_orf1ab.xlsx")
```

```python
data
```

```python
# Prepare the sheet for esm1b input 
df = data.drop_duplicates(subset=['POS_Gene', 'POS', 'REF', 'ALT', 'CDNA_POS', 'CDS_POS',
       'AA_POS', 'Product', 'Start', 'End'])
```

```python
df
```

```python
import pandas as pd

# Assuming you have a DataFrame called 'df'
# Replace 'col2' with the actual column name you want to check
column_name = 'WT_aa'

# Check for spaces in the specified column
has_spaces = df[column_name].str.contains(' ').any()

if has_spaces:
    print(f"There are spaces in '{column_name}' column.")
else:
    print(f"There are no spaces in '{column_name}' column.")
```

```python
df.columns
```

```python
data = df.rename(columns={'POS_Gene': 'Seq_id'})
```

```python
data
```

```python
# Keep only seqId and WT seq for esm1b input 
selected_columns  = ["Seq_id","WT_aa"]
data1 = data[selected_columns]
```

```python
## Esm1B format file 
import pandas as pd

# Output file name
output_file = '../esm1b_analysis/esm1b_pilot_input.fasta'

# Write to the text file
with open(output_file, 'w') as f:
    for index, row in data1.iterrows():
        #f.write(f'>{row["seq_id"]}\n{row["aa_seq_Mut"]}\n')
        f.write(f'>{row["Seq_id"]}\n{row["WT_aa"]}\n')

print('Data written to', output_file)
```
