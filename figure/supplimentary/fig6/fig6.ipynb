```python
import pandas as pd
import matplotlib.pyplot as plt

# Data
data = {
    'Gene': ['ORF1ab', 'S', 'N', 'M', 'E', 'ORF7a', 'ORF3a', 'ORF8', 'ORF7b', 'ORF10', 'ORF6'],
    'Number of unique mutations': [43849, 12582, 3203, 1817, 563, 2539, 2330, 1521, 1082, 789, 670]
}

# Create DataFrame
df = pd.DataFrame(data)

# Plot
plt.figure(dpi=300, figsize=(3, 3))
plt.bar(df['Gene'], df['Number of unique mutations'], color='lightcoral')
plt.xlabel('Genes')
plt.ylabel('Unique utation counts')
plt.title('Number of Unique Mutations in COVID-19 Genes')
plt.xticks(rotation=45, ha='right')
#plt.tight_layout()
#Remove top and right spines,
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
# Remove background lines
plt.grid(False)
plt.show()
#plt.savefig("12.pdf", dpi =600)
```

```python
# file_path = os.path.join(BASE_DIR, "lineages_count.xlsx")

# Read data from Excel file
df = pd.read_excel("../../../data/lineages_count.xlsx")

# Count the number of variants in each lineage
variant_counts = df['Lineage'].value_counts().sort_index()

# Define custom colors (ensure it covers all categories)
custom_colors = ['skyblue', 'green', 'salmon', 'magenta', 'purple', 'gold', 'cyan', 'red']
custom_colors = custom_colors[:len(variant_counts)]  # Match color count with unique lineages

# Create the pie chart
plt.figure(figsize=(8, 8))
wedges, texts, autotexts = plt.pie(
    variant_counts.values, 
    labels=[f"{label} ({count})" for label, count in zip(variant_counts.index, variant_counts.values)],  
    startangle=140, 
    colors=custom_colors, 
    autopct='%1.1f%%',  # Display percentage
    textprops={'fontsize': 8}  # Adjust text size for readability
)

plt.title('Variant Counts in Each Lineage', fontsize=12)
plt.axis('equal')  # Ensure pie chart is circular

# # Save the figure with a relative path
# output_file = os.path.join(BASE_DIR, "variant_counts_pie_chart.pdf")
# plt.savefig(output_file, dpi=600)

# Show the plot
plt.show()
```
