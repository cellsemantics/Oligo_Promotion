##### import library

```python
import fcgr
import pandas 
import kaos
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import utility
```

##### Read fasta

```python
file_path_yeast = "../../../data/w303_vlte.fasta"
file_path_human = "../../../data/GRCh38_latest_genomic.fna"
file_path_ecoli = "../../../data/562.5708.fna"
```

##### Set kmer length

```python
kmer_length = 10
```

##### fig 1b. Heatmap of $\it{H.sapien}$

```python
print("length of sequence of human is: ",  len(kaos.read_fasta(file_path_human)))
fcgr_matrix_human = kaos.chaos_frequency_matrix(fasta_string= kaos.read_fasta(file_path_human), kmer_length = kmer_length,  pseudo_count = True)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(1.33, 1.2), dpi = 600)
heatmap_human =sns.heatmap(-np.log(fcgr_matrix_human[0]/fcgr_matrix_human[0].sum()), ax=ax, cmap= "inferno", cbar_kws={'orientation': 'vertical', 'pad': 0.02})
ax = utility.custom_figure_axis(ax=ax, show_ticks = False, fontsize=5)
cbar1 = heatmap_human.collections[0].colorbar
cbar1.ax.yaxis.set_tick_params(labelsize=5)
title = r"$\it{H.sapien}$ DNA FCGR heatmap"
ax.set_title(title, fontsize = 4)
plt.tight_layout()
```
